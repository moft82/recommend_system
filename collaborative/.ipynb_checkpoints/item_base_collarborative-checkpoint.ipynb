```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
path = 'data'
os.listdir(path)
```

```python
rate = pd.read_csv(os.path.join(path, 'ratings.csv'))
movie = pd.read_csv(os.path.join(path, 'movies.csv'))
print(rate.shape)
print(movie.shape)
```

```python
movie = movie[movie['movieId'].isin(rate['movieId'])]
print(rate.shape)
```

```python
rate
```

```python
movie.reset_index(inplace=True, drop=True)
movie.reset_index(inplace=True)
movie.rename(columns={"index":'id'}, inplace=True)
df = pd.merge(rate, movie[['id', 'movieId']], on='movieId', how='inner')
# movie.drop('movieId', axis=1, inplace=True)
```

```python
pd.merge(rate, movie[['id', 'movieId']], on='movieId', how='inner')
```

```python
pivot = df.pivot(columns='userId', index='id', values='rating')
pivot.fillna(0, inplace=True)
pivot
```

```python
item_base_coll = cosine_similarity(pivot)
item_base_coll
```

```python
rec = item_base_coll[0].argsort()[::-1][1:11]
```

```python
movie_name = movie.loc[rec, 'title'].values
movie_name
```

```python
print("\n".join(movie_name[:5]))
```

```python
def recommender(item_id):
    item = movie.loc[item_id, 'title']
    print("Selected Movie : {}".format(item))
    print("--"*40)
    recommend = item_base_coll[item_id].argsort()[::-1][1:11]
    movie_name = movie.loc[recommend, 'title'].values
    print("\n".join(movie_name[:10]))
```

```python
recommender(0)
```
