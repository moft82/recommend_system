```python
import pandas as pd
import numpy as np
import os
```

```python
path = "data"
f = os.listdir(path)[0]
f
```

```python
df = pd.read_csv(os.path.join(path, f))
print(f"Row : {df.shape[0]}, \t Columns : {df.shape[1]}")
df.head(3)
```

```python
drop_columns = ['master_seq', 'vol', 'author', 'publisher', 'pub_date', 'add_code', 'price', 
         'img_url', 'author_replace', 'pub_date_2','is_coll_aladin', 'is_coll_naver', 'isbn_origin', 'title_replace']
df.drop(drop_columns, axis=1, inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0, inplace=True)
print(df.shape)
df.isna().sum()
```

```python
df.head()
```

```python
df['kdc_class_no'] = df['kdc_class_no'].astype(float)
df['main_class'] = df['kdc_class_no'] // 100
df['sub_class'] = (df['kdc_class_no'] - (df['main_class'] * 100)) // 10
df.head()
```

```python
df = df[(df['main_class'] == 0) & (df['sub_class'] == 0)]
drop_columns = ['kdc_class_no', 'main_class', 'sub_class']
df.drop(drop_columns, axis=1, inplace=True)
df.head()
```

```python
df.to_csv(os.path.join(path, "target.csv"), index=False)
```
